```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
parent_dir = Path('../../likelihood_results/likelihoodv3')

likelihood_dir = parent_dir/'3integrand_diff_10interp_1000'
traj_dir = likelihood_dir/'trajectories'
traj_ids = [1,100,250,500,650,750,1000]
for traj_id in traj_ids:
    traj_csv = traj_dir/f'trajectory_{traj_id}.csv'
    
    traj_df = pd.read_csv(traj_csv,index_col='Index')
    
    t = traj_df['Timestep']
    x = traj_df['Sample']

    print(len(t))
    plt.title('Diffusion (interpolated x10)')
    # plt.title('Diffusion')
    plt.scatter(t,x,s=0.1)
```
